```
import torch
import json 
from transformers import T5Tokenizer, T5ForConditionalGeneration, T5Config
#import pdfminer
import os
import io
import requests
from bs4 import BeautifulSoup as bs
from nltk.tokenize import word_tokenize
```

```
def pre_process_text(text):
  preprocess_text = text.strip().replace("\n","")
  t5_prepared_Text = "summarize: "+preprocess_text
  #print ("original text preprocessed: \n", preprocess_text)
  return preprocess_text,t5_prepared_Text
```

```
def get_summary(t5_prepared_Text):
  tokenized_text = tokenizer.encode(t5_prepared_Text, return_tensors="pt").to(device)
  
  # summmarize 
  summary_ids = model.generate(tokenized_text,
                                      num_beams=4,
                                      no_repeat_ngram_size=2,
                                      min_length=30,
                                      max_length=100,
                                      early_stopping=True)
  output = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
  return output
```

```
#using the t5-base pretrained checkpoint trained on 220 million parameters
model = T5ForConditionalGeneration.from_pretrained('t5-small')
tokenizer = T5Tokenizer.from_pretrained('t5-small')
device = torch.device('cpu')
```

```

def returnWebText(url):
  r = requests.get(url)
  soup = bs(r.text, 'html.parser')
  text = []
  page_content = soup.find_all('p')
  for row in page_content:
    text.append(row.get_text().replace('\n', ' ').strip())
    #print(row.get_text())
    #print(type(row.get_text()))
  return text
  
```

```
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```
import nltk
nltk.download('punkt')
```

```
def main(url):
  p_string = ' '.join(returnWebText(url=url))
  tokens = word_tokenize(p_string)
  print("++++++++++++++++Tokens+++++++++++++++++")
  print(tokens)
  print("++++++++++++++++Chunks+++++++++++++++++")
  chunk = list(chunks(tokens, 1000))
  print(chunk)
  summary = []
  for values in chunk:
    _ , prepared_text = pre_process_text(' '.join(values))
    partial_summary = get_summary(prepared_text)
    summary.append(partial_summary)
    print("+++++++++++++++++++Prepared Text++++++++++++++++")
    print(prepared_text)
    print("+++++++++++++++++++Summary++++++++++++++++++++++")
    print(partial_summary)

  
  print(' '.join(summary))
  with open(os.path.abspath('.') + '/Summary_' + url.split('.')[1] + '.txt', 'w' ) as f:
        f.write("Summary of " + url +":\n\n" )
        f.write("\n\n===========================================\n\n" )
        f.write(' '.join(summary))
    


```
